```python
import pandas as pd
```

```python
df = pd.read_csv("/content/weather_data.csv", parse_dates=["day"])
df.set_index("day", inplace=True)
df
```

```python
df_all_zeros = df.fillna(0)
df_all_zeros
```

```python
df_custom_filled_nans = df.fillna({"temperature": df.temperature.mean(), "windspeed": df.windspeed.median(), "event": "No event"})
df_custom_filled_nans
```

```python
df.fillna(method="ffill")
```

```python
df.fillna(method="bfill")
```

```python
df.interpolate()
```

```python
df.dropna()
```

```python
df.dropna(how="all")
```

```python
df.dropna(thresh=1)
```
